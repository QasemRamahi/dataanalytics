```python
categorical_columns = ['behaviour', 'Type']

categorical_summary = {}

for column in categorical_columns:
    value_counts = df[column].value_counts()
    categorical_summary[column] = {
        'count': value_counts.sum(),
        'unique': len(value_counts),
        'top': value_counts.idxmax(),
        'freq': value_counts.max()
    }

summary_df = pd.DataFrame(categorical_summary).T

print(summary_df)
```

```python
df_normal_node =df[df['Type'] == 'normal']
df_normal_node = df_normal_node[df_normal_node["Node_id"]== 80]
df_normal_node2 =df[df['Type'] == 'normal']
df_normal_node2 = df_normal_node2[df_normal_node2["Node_id"]== 83]
df_normal_node.head(50)
```

```python
#this is the first pattern, most forwarding attacks are on the hop count betweeen 7 and 14

plt.figure(figsize=(10, 6))


attack_F = dflast
attack_N = df_normal_node

plt.plot(attack_F['Hop_Count'], attack_F['Type'], marker='o', label='Attack: Forwarding')

plt.plot(attack_N['Hop_Count'], attack_N['Type'], marker='s', label='no attack: normal')

plt.xlabel('Hop Count')
plt.ylabel('type')
plt.title('Hop Count with Attack Types')

plt.legend()
plt.grid(True)
plt.show()

```
